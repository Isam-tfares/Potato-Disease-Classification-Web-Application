<h3>importing the necessary libraries</h3>

```python
import numpy as np
import pandas as pd
import cv2
import os

import matplotlib.pyplot as plt
import seaborn as sns
import random

from imblearn.over_sampling import SMOTE

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import load_model

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
# import joblib
# import pickle
from keras.models import load_model
```

<h3>load and transforming data</h3>

```python
os.listdir('./PlantVillage/')
```

```python
classes=os.listdir('./PlantVillage/')
classes
```

```python
X=[]
y=[]
for i,folder in enumerate(classes):
    for image_name in os.listdir('./PlantVillage/'+folder):
        image_path='./PlantVillage/'+folder+'/'+image_name
        img=cv2.imread(image_path)
        img=cv2.resize(img,(128,128))
        X.append(img)
        y.append(i)
```

```python
X=np.array(X)
y=np.array(y)
```

<h3>Shuffle data</h3>

```python
shuffler = np.random.permutation(len(X))
X = X[shuffler]
y = y[shuffler]
```

```python
X.shape
```

<h3>plot some images</h3>

```python
plt.figure(figsize=(10, 10))
for i in range(12):
    index=random.randint(0, 2151)
    ax = plt.subplot(3, 4, i + 1)
    plt.imshow(X[index])
    plt.title(classes[y[index]])
    plt.axis("off")
```

<h3>We can observe that Potato___healthy doesn't exist</h3>

```python
for classname in classes:
    print(classname,len(os.listdir('./PlantVillage/'+classname)))
```

<h3>So data is unbalanced</h3>

<h3>So we use SMOTE to balance data</h3>

```python
smote = SMOTE(sampling_strategy='auto') 
```

<h3>First we should reshape data to 2D array </h3>

```python
X=X.reshape(X.shape[0],-1)
```

```python
X, y = smote.fit_resample(X, y)
```

```python
X=X.reshape((3000,128,128,3))
```

```python
X.shape
```

```python
plt.figure(figsize=(10, 10))
for i in range(12):
    index=random.randint(0, 3000)
    ax = plt.subplot(3, 4, i + 1)
    plt.imshow(X[index])
    plt.title(classes[y[index]])
    plt.axis("off")
```

<h3>transform y to categorical</h3>

```python
y = tf.keras.utils.to_categorical(y, num_classes=len(classes))
```

<h3>Scalling data and split data</h3>

```python
X=X/255.0
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2)
X_test,X_val,y_test,y_val=train_test_split(X_test,y_test,test_size=.4)
```

```python
X_train.shape,X_test.shape,X_val.shape
```

<h3>Data augmentation </h3>

```python
data_augmentation = keras.Sequential([
    keras.layers.RandomFlip("horizontal_and_vertical"),
    keras.layers.RandomRotation(0.2),
    keras.layers.RandomZoom(0.2),  # Add RandomZoom for random zooming
    keras.layers.RandomContrast(0.2),  # Add RandomContrast for adjusting image contrast
])
```

```python
plt.figure(figsize=(10, 10))
for i in range(9):
      index=random.randint(0, 2400)
      augmented_image = data_augmentation(X_train[index])
      ax = plt.subplot(3, 3, i + 1)
#       plt.imshow(augmented_image)
      plt.imshow(np.array(augmented_image * 255).astype(np.uint8))
      plt.axis("off")
```

```python
augmented_data=data_augmentation(X_train)
```

<h3>Building model</h3>

```python
input_shape=(128, 128, 3)
model = keras.Sequential([
  # Features extraction
  keras.layers.Conv2D(16, 3, padding='same', activation='relu',input_shape=input_shape),
  keras.layers.MaxPooling2D(),
  keras.layers.Conv2D(64,  kernel_size = (3,3), activation='relu'),
  keras.layers.MaxPooling2D((2, 2)),
  keras.layers.Conv2D(64,  kernel_size = (3,3), activation='relu'),
  keras.layers.MaxPooling2D((2, 2)),
  # classification ANN
  keras.layers.Flatten(),
  keras.layers.Dense(32,activation='relu'),
  keras.layers.Dense(len(classes),activation="softmax")
])
model.build()
```

```python
model.summary()
```

```python
model.compile(
    optimizer='SGD',
    loss=keras.losses.categorical_crossentropy,
    metrics=['accuracy']
)
```

```python
EPOCHS=100
history=model.fit(augmented_data,y_train,validation_data=(X_val,y_val),epochs=EPOCHS)
```

```python
model.evaluate(X_test,y_test)
```

```python
history.history.keys()
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']
plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(range(EPOCHS), acc, label='Training Accuracy')
plt.plot(range(EPOCHS), val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(range(EPOCHS), loss, label='Training Loss')
plt.plot(range(EPOCHS), val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

```python
def predict(model, img,is_scalled=True):
    if not is_scalled :
        img=img/255.0
        
    predictions = model.predict(img)
    predicted_class = classes[np.argmax(predictions[0])]
    confidence = round(100 * (np.max(predictions[0])), 2)
    return predicted_class, confidence
```

```python
predict(model,X_test[:1])
```

```python
plt.figure(figsize=(10, 10))
for i in range(9):
    index=random.randint(0, 359)
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(np.array(X_test[index] * 255).astype(np.uint8))
    predicted_class, confidence=predict(model,X_test[index:index+1])
    plt.title(predicted_class+' '+str(confidence)+'%')
    plt.axis("off")
```
